Se pide crear una aplicación que actúe como un buscador con varias opciones:
 
 * __Opción 1__: Buscar por nombre de película. El usuario podrá introducir el nombre de una película y se deberá mostrar de forma amigable lo recuperado por cada película reconocida.
 
 * __Opción 2__: Buscar por identificador de película. El usuario podrá introducir elidentificador de una película y se deberá mostrar de forma amigable lo recuperado por cada película reconocida.
 
 * __Opción 3__: Búsqueda ampliada. Es una combinación de las opciones 1 y 2.El usuario introduce el nombre de una película y por cada película recuperada se deberá mostrar la información recuperada más la información adicional que se puede recuperar de cada película usando el identificador de pelicula. 
 
 * __Opción 4__: Recuperacion de posters.El usuario podrá introducir el nombre de una película y se deberá mostrar por pantalla el nombre de cada película recuperada y la dirección al poster. Además se creará una carpeta donde se almacenarán los posters que deberán ser recuperados a partir de las direcciones indicadas.

 * __Opción 5__: Recuperacion con poster.El usuario podrá introducir el nombre de una película y se deberá mostrar por pantalla los nombres de las diferentes películas recuperadas. A continuación, el usuario elegirá una de esas películas y entonces se le mostrará por pantalla toda la información ampliada sobre esa película (como en la opción 2) pero además se le mostrará el poster de la película (para ello hay que bajarse previamente el poster de la película y usar el método Image como se realiza en los apuntes.
 * __Opción 6__: Salir o continuar.


```python
import urllib.request
from urllib.parse import urlencode, quote_plus
clave="849edfa5"

import json
import os
from IPython.display import Image, display


def buscaNombre(nombre):
    uri="http://www.omdbapi.com/?apikey="+clave
    url = uri + "&"+urlencode({"s":nombre, "type":"movie"},quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    info= []
    for line in manf:
        PythonDatos=json.loads(line.decode())
    
    return PythonDatos


def imprimeNombre(PythonDatos):
    for movie in PythonDatos["Search"]:
        for data in movie.keys():
            print(data + ": " + movie[data])
        print("\n")
    
    
def buscaID(id):
    uri="http://www.omdbapi.com/?apikey="+clave
    url = uri + "&"+urlencode({"i":id},quote_via=quote_plus)
    manf = urllib.request.urlopen(url)
    info= []
    for line in manf:
        PythonDatos=json.loads(line.decode())

    for data in PythonDatos.keys():
        if data != "Ratings": 
            print(data + ": " + PythonDatos[data])
        else:
            print("Ratings: ")
            for rating in PythonDatos[data]:
                for key in rating.keys():
                    print(key + ": "+ rating[key])
                print("\n")
    print("\n")


def buscaDoble(nombre): 
    datos = buscaNombre(nombre)
    
    for movie in datos["Search"]:
        buscaID(movie["imdbID"])

def buscaPosters(nombre):
    datos = buscaNombre(nombre)
    
    if not os.path.exists(nombre):
        os.makedirs(nombre)
        
    for movie in datos["Search"]:
        print("Title: " + movie["Title"] +"\nPoster: " + movie["Poster"] + "\n" )
        
        if movie["Poster"] != "N/A":
            img = urllib.request.urlopen(movie["Poster"]).read()
            manf=open(nombre + '/' +  movie["Title"].replace(":", "") + '.jpg',"wb")
            manf.write(img)
            manf.close()

def buscaPoster(nombre):
    datos = buscaNombre(nombre)
    
    index = 0
    listaPelis =[]
    for movie in datos["Search"]:
        print(str(index) + ".- Title: " + movie["Title"] + "\n")
        listaPelis.append([movie["Title"], movie["imdbID"], movie["Poster"]])
        index +=1
     
    seleccion = -1
    
    while(seleccion < 0 or seleccion > index):
         seleccion = int(input("Introduce el numero correspondiente a la pelicula de la que quieras obtener informacion: "))
    
    nombre = listaPelis[seleccion][0].replace(":", "")
    if not os.path.exists(nombre):
        os.makedirs(nombre)
    
    buscaID(listaPelis[seleccion][1])
    if listaPelis[seleccion][2] != "N/A":
        img = urllib.request.urlopen(listaPelis[seleccion][2]).read()
        manf=open(nombre + '/' +  nombre + '.png',"wb")
        manf.write(img)
        manf.close()
        display(Image(filename = nombre + '/' + nombre + ".png"))
    else: print("No se ha encontrado poster para esta pelicula\n")

def menu():
    while True:
        seleccion = input("¿Que deseas hacer? :\n1.-Busqueda por nombre\n2.-Busqueda por id\n" 
                          + "3.-Busqueda ampliada\n4.-Recuperacion de posters\n"+
                          "5.-Recuperacion con poster\n6.-Salir" )
        if(seleccion == "1"):
            nombre = input("Introduce el nombre de la pelicula a buscar:\n")
            imprimeNombre(buscaNombre(nombre))
        elif(seleccion == "2"):
            ID =input("Introduce el ID de la pelicula a buscar:\n")
            buscaID(ID)
        elif(seleccion == "3"):
            nombre = input("Introduce el nombre de la pelicula a buscar:\n")
            buscaDoble(nombre)
        elif(seleccion =="4"):
            nombre = input("Introduce el nombre de la pelicula a buscar:\n")
            buscaPosters(nombre)
        elif(seleccion == "5"):
            nombre = input("Introduce el nombre de la pelicula a buscar:\n")
            buscaPoster(nombre)
        elif(seleccion == "6"):
            break

menu()
```
